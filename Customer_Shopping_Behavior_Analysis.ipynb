# Handling the null values

### Size - column

```python
# Size
df['Size'].unique()
```

```python
df.head(3)
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# handleing null values or replacing all values in the Electronics category

df.loc[df['Category'] == 'Electronics', 'Size'] = 'Not Applicable' # Electronics
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# handleing null values or replacing all values in the Accessories category

df.loc[df['Category'] == 'Accessories', 'Size'] = 'Free Size' #Accessories
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# handleing null values or replacing all values in the Footware category

df.loc[df['Category'] == 'Footwear', 'Size'] = 'Not Available' #Footwear
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.isnull().sum() 
```

```python
# handleing null values or replacing all values in the Clothing category

df.loc[(df['Category'] == 'Clothing') & (df['Size'].isnull()), "Size"] = df[df["Category"] == "Clothing"]["Size"].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.isnull().sum() 
```

```python
df.head(3)
```

### Review Rating - column

```python
# mean of Review Rating
df['Review Rating'].mean()
```

```python
# median of Review Rating
df['Review Rating'].median()
```

```python
# mode of Review Rating
df['Review Rating'].mode()
```

```python
df.groupby("Item Purchased")["Review Rating"].mean()
```

```python
# Fillna null values of Review Rating with the mean of each Item Purchased

df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')["Review Rating"].transform('mean'))
```

```python
df.isnull().sum() 
```

```python
df.head()
```

```python
df['Review Rating'] = df['Review Rating'].round(2)
```

```python
df.head()
```

### Previous Purchases - column

```python
## Previous Purchases
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum() 
```

### Purchase Amount (USD) - column

```python
### Purchase Amount (USD)
df.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

```python
# Fillna null values of Purchase Amount (USD) with the mean of each Item Purchased
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')["Purchase Amount (USD)"].transform('mean'))
```
